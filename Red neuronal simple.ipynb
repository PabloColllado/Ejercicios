```python
#En primer lugar se importará las librerías necesarias
#Tensorflow es una libreria para inteligencia artificial
import tensorflow.compat.v1 as tf
#Numpy para trabajar facilmente con arreglos numericos
import numpy as np
```

```python
#Ejemplos que la red utilizará para aprender
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

```python
#Con el framework KERAS nos permite hacer las redes neuronales de una forma muy simple
#Crearemos una variable llamada capa densa, que son aquellas que tienen conexiones con todas las neuronas de la siguientes capas
capa = tf.keras.layers.Dense(units=1, input_shape=[1])
modelo = tf.keras.Sequential([capa])
```

```python
#Compilar el modelo para ser entrenado
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss="mean_squared_error"
)
```

```python
print("Comenzando entrenamiento...")
historial = modelo.fit(celsius, fahrenheit, epochs=1000, verbose=False)
print("Modelo entrenado!")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pérdida")
plt.plot(historial.history["loss"])
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="tensorflow")
print("Hagamos una predicción")
resultado = modelo.predict([100.0])
print("El resultado es " + str(resultado) + "fahrenheit")
```

```python
print("variables internas del modelo")
print(capa.get_weights())
```
